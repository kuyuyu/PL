```python
import pandas as pd
eq = pd.read_csv("earthquake202301.csv", encoding='big5')
eq
```

```python
#時間有誤...
#列出發生於凌晨零點至早上六點之間的地震資料

eq['地震時間'] = pd.to_datetime(eq['地震時間'],format='%H:%M')

#dt.hour是日期時間列的一個屬性，它提取日期時間值的小時部分
time_0to6 = eq[(eq['地震時間'].dt.hour >= 0) & (eq['地震時間'].dt.hour < 6)]
occur_time_0to6 = len(time_0to6)

print('發生於凌晨零點至早上六點之間的地震共',occur_time_0to6,'筆')
time_0to6
```

```python
#時間有誤...
#找出一天內發生兩起或以上且時間相同的地震紀錄，即找出地震日期欄位內相同且同時地震時間欄位內相同的資料

eq['地震日期'] = pd.to_datetime(eq['地震日期'], format='%Y年%m月%d日')
eq['地震時間'] = pd.to_datetime(eq['地震時間'], format='%H:%M')

DT = eq.groupby(['地震日期', '地震時間']).size()
duplicates = DT[DT >= 2] #找出重複的地震資料
#set_index將原始地震資料的索引設置為日期和時間的多重索引
#isin將重複出現過的日期和時間的索引與原始地震資料的索引進行比較，從而找出重複的地震資料
result = eq[eq.set_index(['地震日期', '地震時間']).index.isin(duplicates.index)]
result
```

```python
#eq['地震日期']，print出來會有問題...
#嘗試一：列出發生最多地震的一天，並且顯示共發生幾次

eq['地震日期'] = pd.to_datetime(eq['地震日期'],format='%Y年%m月%d日') #output和打的'%Y年%m月%d日'不一樣!
#將地震日期的欄位轉換為 Pandas 的日期時間格式，使用pd.to_datetime()
#日期欄位內注意格式 format程式碼需打出年月日
eq_count = eq.groupby('地震日期').size()
#groupby對日期進行分組，將相同日期的地震事件放在同一個組裡
#size 方法計算每個組的大小，即每天發生地震的次數
most_eqs_date = eq_count.idxmax()
#idxmax會返回序列中值最大的元素對應的索引 (index)

print('發生地震最多的日期為：', most_eqs_date.date(), '，共發生了', eq_count[most_eqs_date], '次地震')
```

```python
#錯誤...
#嘗試二：列出發生最多地震的一天，並且顯示共發生幾次

series_date = eq['地震日期']
datedata = pd.Series(series_date.astype(str))  # Convert to string format

grouped = datedata.str.extract('(\d+)').astype(int).groupby(0).size()
#grouped = datedata.groupby(datedata.str.extract('(\d+)').astype(int)).size()

most_eqs_date_str = grouped.idxmax()
most_eqs_date = pd.to_datetime(most_eqs_date_str)

print('發生地震最多的日期為：', most_eqs_date.strftime('%Y年%m月%d日'), '，共發生了', grouped[most_eqs_date_str], '次地震')
#print('發生地震最多的日期為：', most_eqs_date.date(), '，共發生了', grouped[most_eqs_date_str], '次地震')
```
