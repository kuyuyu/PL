```python
import pandas as pd

#資料源自中央氣象局地震測報中心，載取2023年2月份資料
eq02 = pd.read_csv("eq202302.csv", encoding='big5')
eq02
```

```python
#加問一
#列出在2023年2月極淺層地震中所有芮氏規模大於等於5的資料

depth_scale = eq02[(eq02['深度'] < 30) & (eq02['規模'] >= 5)]
d_s_sum = len(depth_scale)
print('芮氏規模大於等於5且深度小於30km的地震共', d_s_sum,'筆')
depth_scale
```

```python
#加問二
#找出發生最多地震的縣市

#str.slice(stop=3)表只取位置欄位中的前三字
city = eq02['位置'].str.slice(stop=3)
#string不支持groupby，因此轉換成series
city_series = pd.Series(city)
#將各縣市發生的次數分組並計算大小
city_counts = city_series.groupby(city_series).size()
#value_counts函數來統計每個縣市出現的次數
most_freq_city = city.value_counts().idxmax()
#將發生地震次數最多的縣市存在city_counts的series，並宣告成happened_count的變數
happened_count = city_counts[most_freq_city]

print('發生地震最多的縣市為：', most_freq_city,'共',happened_count,'筆')
```

```python
#加問三
#找出地震發生最多的一天

eq02_datetime = pd.to_datetime(eq02['地震時間'], format='%Y/%m/%d %H:%M')

# 計算每一天地震發生的次數
eq02_count = eq02_datetime.groupby(eq02_datetime.dt.date).size()

# 找出地震發生次數最多的日期
most_freq_date = eq02_count.idxmax()

print('地震發生最多的日期為：', most_freq_date, '，共發生了', eq02_count[most_freq_date], '次地震')
```
