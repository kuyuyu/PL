```python
#ex.https://www.ptt.cc/bbs/PublicServan/index6.html
htmlURL = "https://www.ptt.cc/bbs/PublicServan/index"
htmlURL
```

```python
page = range(1,11)
urlfirst = "https://www.ptt.cc/"
allurl = list()
for i in page:
    link = htmlURL + str(i) + ".html"
    response = requests.get(link)
    response.encoding = "utf-8"
    soup = BeautifulSoup(response.text, "html.parser")
    a_tags = soup.find_all("a")
    for a in a_tags:
        href = a.get("href")
        if href and href.startswith("/"):
            allurl.append(urlfirst + href)
```
