```python
import requests
import time
from time import sleep
from bs4 import BeautifulSoup
import csv
```

```python
head = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'
}
url = 'https://www.greenpeace.org/taiwan/%e5%8f%83%e8%88%87%e7%92%b0%e5%a2%83%e4%bf%9d%e8%ad%b7%e7%9a%84%e6%96%b9%e5%bc%8f/'
response = requests.get(url, headers=head)
if response.status_code == 200:
    print('請求成功！')
    soup = BeautifulSoup(response.text, 'html.parser')
else:
    print('請求失敗，狀態碼為 {response.status_code}。')
```

```python
#備用
titles = soup.find_all('h3', class_='title')
urls = soup.find_all('a', class_='swiper-slide')

for title, url in zip(titles, urls):
    print('活動名稱:', title.text.strip())
    print('活動網址:', url['href'])
```

```python
event_titles = soup.find_all('h3', class_='title')
event_subtitles = soup.find_all('h4', class_='subtitle')
event_urls = soup.find_all('a', class_='swiper-slide')

for event_title, event_subtitle, event_url in zip(event_titles, event_subtitles, event_urls):
    subtitle_text = event_subtitle.text.strip()
    subtitle_text = subtitle_text.replace('目標：', '')
    print('活動名稱:', subtitle_text)
    print('活動網址:', event_url['href'])
```

```python
with open('consumer.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['活動名稱', '活動網址'])
    
    for event_title, event_subtitle, event_url in zip(event_titles, event_subtitles, event_urls):
        subtitle_text = event_subtitle.text.strip()
        subtitle_text = subtitle_text.replace('目標：', '')
        writer.writerow([subtitle_text, event_url['href']])
```
