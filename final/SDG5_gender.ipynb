```python
import requests
import time
from time import sleep
from bs4 import BeautifulSoup
import csv
```

```python
head = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'
}
response = requests.get('https://tgeea.org.tw/event/', headers=head)
if response.status_code == 200:
    print('請求成功！')
    soup = BeautifulSoup(response.text, 'html.parser')
else:
    print('請求失敗，狀態碼為 {response.status_code}。')
```

```python
event_titles = soup.find_all('h3', class_='elementor-post__title')

for event_title in event_titles:
    event_name = event_title.find('a').text.strip()
    event_url = event_title.find('a')['href']
    print('活動名稱:', event_name)
    print('活動網址:', event_url)
```

```python
with open('gender.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['活動名稱', '活動網址'])

    for event_title in event_titles:
        event_name = event_title.find('a').text.strip()
        event_url = event_title.find('a')['href']
        writer.writerow([event_name, event_url])
```
