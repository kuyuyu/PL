```python
#找出魔鏡歌詞網中有列出的全部理想混蛋的歌的網址
link = []

for i in range(1, 21):
    if i == 9:
        link.append('https://mojim.com/twy219017x9x1.htm')
        link.append('https://mojim.com/twy219017x9x2.htm')
    elif i == 11:
        for j in range(1, 12):
            link.append(f'https://mojim.com/twy219017x{i}x{j}.htm')
    elif i == 18:
        for j in range(1, 11):
            link.append(f'https://mojim.com/twy219017x{i}x{j}.htm')
    elif i != 2:
        link.append(f'https://mojim.com/twy219017x{i}x1.htm')
```

```python
all_lyrics = []  # 用來儲存所有歌詞

for url in link:
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    tag_dl = soup.find_all('dd', {'id': 'fsZx3', 'class': 'fsZx3'})
    lyrics = []  # 用來儲存當前網址的歌詞
    br_count = 0  # 計算出現連續三個<br>的次數

    for dl in tag_dl:
        tag_br = dl.find_all('br')
        for br in tag_br:
            next_tag = br.next_sibling
            if next_tag is not None and next_tag.name != 'br':
                lyrics.append(next_tag.strip())
            else:
                br_count += 1
                if br_count == 10:
                    all_lyrics.append(lyrics)
                    break

        if br_count == 10:
            # 已經抓取完當前網址的所有歌詞，跳出迴圈
            break

all_lyrics
```

```python
def get_lyrics(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    lyrics = []    
    for div in soup.find_all('div', {'id': 'fsZ', 'class': 'fsZ'}):
        for dl in div.find_all('dl', {'id': 'fsZx1', 'class': 'fsZx1'}):
            for br in dl.find_all('br'):
                lyrics.append(br.next_sibling)
    return lyrics

for url in link:
    lyrics = get_lyrics(url)
    print(lyrics)
```

```python
lyrics_list = []
for url in link:
    lyrics = get_lyrics(url)
    lyrics_list.append(lyrics)

print(f"共抓到 {len(lyrics_list)} 首歌的歌詞。")
```
