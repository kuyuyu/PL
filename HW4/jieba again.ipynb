```python
import pandas as pd

bestard = pd.read_csv('bestards自造4.csv', encoding='big5')
bestard.head()
```

```python
import jieba
import jieba.analyse
import numpy
from collections import Counter
```

```python
documents = ['理想混蛋是真的很好聽，超級推薦大家去聽！']

for sentence in documents:
    seg_list = jieba.cut(sentence)
    print('/'.join(seg_list))
```

```python
segments = []
for comment in bestard['歌名與歌詞']:
    seg_list = jieba.lcut(comment)
    segments.append(seg_list)

# 把 list 轉成 DataFrame
df_segments = pd.DataFrame({'seg': segments})

# 顯示斷詞結果
print(df_segments.head())
```

```python
#合成list
LIST = df_segments['seg'].to_list()
#print(LIST)
```

```python
#將kist內部的元素拆分
List = []
for key in range(len(LIST)):
    for value in LIST[key]:
        List.append(value)
#print(List)
```

```python
results = pd.value_counts(List)
results.head(50)
```

```python
bestard['seg'] = ""
bestard.head()
```

```python
index = 0
for documents in bestard['歌名與歌詞'].astype(str):
    #seg_list = jieba.analyse.extract_tags(documents, 150)
    seg_list = jieba.lcut(documents)
    bestard['seg'].iloc[index] = seg_list
    index += 1
```

```python
words = list(bestard['歌名與歌詞'])
show_key = pd.DataFrame(Counter(words).keys())
show_values = pd.DataFrame(Counter(words).values())
df = [show_key, show_values]
```

```python
result = pd.concat(df, axis=1)
result.columns = ['word', 'frequency']
```

```python
import plotly.express as px

px.bar(result, x='word', y='frequency', title='amount of lyrics')
```
